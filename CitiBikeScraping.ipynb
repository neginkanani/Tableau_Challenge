```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
import urllib.request 
import zipfile
```

```python

'''Scrape the citibike website to collect the zipfiles'''
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
#reading the html code
html = browser.html
soup = BeautifulSoup(html, "lxml")

url_citi = 'https://s3.amazonaws.com/tripdata/index.html'
browser.visit(url_citi)
time.sleep(1)


citibike_data = []
for j in range(220):
    citibike = browser.find_by_xpath('//*[@id="tbody-content"]/tr'+ str([j]) +'/td[1]/a[@href]')
    for i in citibike:
        title_link = i.text
        citibike_data.append(title_link)
        time.sleep(2)


browser.quit()
citibike_data
```

```python
#Downloading the zipfiles in the same directory as the code here
for j in range(220):
    url = "https://s3.amazonaws.com/tripdata/" + citibike_data [j]
    filename = citibike_data [j]
    urllib.request.urlretrieve(url, filename) 
 
print("Download complete.")
```

```python
# read the zip file first we need to get the name of the file without.zip or .zip.csv for the next step
file_name=[]
for j in citibike_data:
    file_name.append(j.split(".")[0])
file_name
```

```python
#Now read the csv inside zipfile and include very 1000 rows
df_all=pd.DataFrame()
for j in citibike_data:
    try:
        zf = zipfile.ZipFile('C:/Users/nkananizadeh/Desktop/Negin/Tableau/'+j) 
        try:
            df = pd.read_csv(zf.open(str(j.split(".")[0])+".csv"))[::1000]
            df_all= pd.concat([df_all,df])
        except:
            print("not a good data:" +str(j.split(".")[0]))
    except:
        print("No such file or directory")
```
